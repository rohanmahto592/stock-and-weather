```python
currentdates=sorted(list(set(df.index.date)))
```

```python
def rates(opn,high,low,close):
    l=len(opn)
    current=1
    prev=0
    shortpos=-1
    short=False
    while current<l:
        if low[current]<low[prev]:
            short=True
            shortpos=current
            break
        prev=current
        current+=1
    if short:
        while current<l:
            if high[shortpos]<high[current]:
                return close[current]-opn[0]
            current+=1
    return close[-1]-opn[0]
            
    
```

```python
for date in currentdates[0:]:
    dataset=df.loc[str(date)]
    opn,high,low,close=[],[],[],[]
    for i in range(dataset.shape[0]):
        opn.append((dataset.iloc[i])[0])
        high.append((dataset.iloc[i])[1])
        low.append((dataset.iloc[i])[2])
        close.append((dataset.iloc[i])[3])
        
    result=rates(opn,high,low,close)
    if result<0:
        print(f"{date}: Loss \t{-round(result,2)}")
    elif result>0:
        print(f"{date}: Profit \t{round(result,2)}")
    else:
        print(f"{date}: No Profit or Loss")
```
